Convert Pdf's into images

```python
import os
from pdf2image import convert_from_path
import pytesseract
from PIL import Image

# Folder containing the PDFs
pdf_folder_path = '/content/drive/MyDrive/Business Contract Validation/Reference Contracts/'

# Folder where images will be saved (inside 'outputs' folder)
output_folder_path = '/content/drive/MyDrive/Business Contract Validation/outputs/'

# Get all PDF files in the folder
pdf_files = [f for f in os.listdir(pdf_folder_path) if f.endswith('.pdf')]

# Loop through each PDF file in the folder
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder_path, pdf_file)
    print(f"Processing file: {pdf_path}")

    # Create a subfolder for the current contract (PDF file name without extension) inside 'outputs'
    contract_name = os.path.splitext(pdf_file)[0]  # Remove file extension
    contract_folder_path = os.path.join(output_folder_path, contract_name)

    # Create the folder for the contract if it doesn't exist
    if not os.path.exists(contract_folder_path):
        os.makedirs(contract_folder_path)

    # Create the 'images' subfolder inside the contract folder
    images_folder_path = os.path.join(contract_folder_path, 'images')
    if not os.path.exists(images_folder_path):
        os.makedirs(images_folder_path)

    # Convert PDF to a list of images
    images = convert_from_path(pdf_path)

    # Loop through each image and save them in the 'images' folder
    for i, image in enumerate(images):
        # Save the image in the 'images' folder inside the contract's folder
        image_path = os.path.join(images_folder_path, f'page_{i+1}.png')
        image.save(image_path, 'PNG')

    print(f"Images saved for {pdf_file}")

# Now, the images for each contract will be saved inside 'images' folders in 'outputs', with each contract having its own folder.
```

Text Extraction

```python
import os
import pytesseract
from PIL import Image

# Folder containing the contracts (already processed images inside the 'images' folder)
output_folder_path = '/content/drive/MyDrive/Business Contract Validation/outputs/'

# Loop through each contract folder in the 'outputs' folder
contract_folders = [f for f in os.listdir(output_folder_path) if os.path.isdir(os.path.join(output_folder_path, f))]

# Loop through each contract folder
for contract_folder in contract_folders:
    contract_folder_path = os.path.join(output_folder_path, contract_folder)
    images_folder_path = os.path.join(contract_folder_path, 'images')

    # Check if the 'images' folder exists
    if os.path.exists(images_folder_path):
        # Create the text file to store all extracted text
        text_file_path = os.path.join(contract_folder_path, f'{contract_folder}_extracted_text.txt')

        # Initialize an empty string to store all extracted text
        extracted_text = ""

        # Loop through each image in the 'images' folder
        image_files = [f for f in os.listdir(images_folder_path) if f.endswith('.png')]

        # Loop through each image and extract text
        for image_file in image_files:
            image_path = os.path.join(images_folder_path, image_file)

            # Open the image
            image = Image.open(image_path)

            # Extract text from the image using pytesseract
            text = pytesseract.image_to_string(image, lang='eng')

            # Append the extracted text to the result
            extracted_text += text + "\n\n"  # Adding a newline between images

        # Save all the extracted text into the text file
        with open(text_file_path, 'w') as text_file:
            text_file.write(extracted_text)

        print(f"Extracted text saved for {contract_folder}")
    else:
        print(f"No 'images' folder found for {contract_folder}")
```

Segmentation

```python
import os
import re

# Folder containing the contracts (already processed images inside the 'images' folder)
output_folder_path = '/content/drive/MyDrive/Business Contract Validation/outputs/'

# Function to segment the text based on a simple heading pattern (you can modify the pattern)
def segment_text(text):
    # Example pattern for headings - modify this to match the actual headings in your contract texts
    heading_pattern = r'\b(Section|Article|Clause|Chapter|Part)\b[\s:]+[\w\s]+'

    # Find all headings based on the pattern
    headings = re.findall(heading_pattern, text)

    # If headings are found, split the text into segments based on the headings
    if headings:
        segments = re.split(heading_pattern, text)
        segments = [segment.strip() for segment in segments if segment.strip()]  # Remove empty segments
    else:
        # If no headings, consider the entire text as one segment
        segments = [text.strip()]

    return segments

# Loop through each contract folder in the 'outputs' folder
contract_folders = [f for f in os.listdir(output_folder_path) if os.path.isdir(os.path.join(output_folder_path, f))]

# Loop through each contract folder
for contract_folder in contract_folders:
    contract_folder_path = os.path.join(output_folder_path, contract_folder)
    extracted_text_file_path = os.path.join(contract_folder_path, f'{contract_folder}_extracted_text.txt')

    # Check if the extracted text file exists
    if os.path.exists(extracted_text_file_path):
        # Read the extracted text from the file
        with open(extracted_text_file_path, 'r') as text_file:
            extracted_text = text_file.read()

        # Segment the extracted text
        segments = segment_text(extracted_text)

        # Create a new file to save the segmented text
        segmented_text_file_path = os.path.join(contract_folder_path, f'{contract_folder}_segmented_text.txt')

        # Save the segmented text into the new file
        with open(segmented_text_file_path, 'w') as segmented_file:
            for idx, segment in enumerate(segments):
                segmented_file.write(f"--- Segment {idx+1} ---\n")
                segmented_file.write(segment + "\n\n")

        print(f"Segmented text saved for {contract_folder}")
    else:
        print(f"Extracted text file not found for {contract_folder}")
```

```python
import os
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import re

def process_input_contract(pdf_path):
    # Step 1: Ask the user for the type of contract
    contract_type = input("Enter the type of contract (e.g., ContractTypeA, ContractTypeB): ").strip()
    if not contract_type:
        raise ValueError("Contract type cannot be empty. Please provide a valid contract type.")

    # Get contract name and create necessary folders in the input contract folder
    contract_name = os.path.splitext(os.path.basename(pdf_path))[0]  # Get file name without extension
    contract_folder_path = os.path.dirname(pdf_path)  # Input contract folder path

    # Create the folder for images if it doesn't exist
    images_folder_path = os.path.join(contract_folder_path, 'images')
    os.makedirs(images_folder_path, exist_ok=True)

    # Step 2: Convert PDF to images
    images = convert_from_path(pdf_path)
    for i, image in enumerate(images):
        image_path = os.path.join(images_folder_path, f'page_{i+1}.png')
        image.save(image_path, 'PNG')

    print(f"Images saved for {pdf_path}")

    # Step 3: Extract text from the images using OCR
    extracted_text = ""
    for image_file in sorted(os.listdir(images_folder_path)):
        image_path = os.path.join(images_folder_path, image_file)
        text = pytesseract.image_to_string(Image.open(image_path), lang='eng')
        extracted_text += text + "\n\n"

    text_file_path = os.path.join(contract_folder_path, f'{contract_name}_extracted_text.txt')
    with open(text_file_path, 'w') as text_file:
        text_file.write(extracted_text)

    print(f"Extracted text saved for {contract_name}")

    # Step 4: Segment the extracted text based on headings (e.g., Section, Article, Clause)
    segments = segment_text(extracted_text)
    segmented_text_file_path = os.path.join(contract_folder_path, f'{contract_name}_segmented_text.txt')
    with open(segmented_text_file_path, 'w') as segmented_file:
        for idx, segment in enumerate(segments):
            segmented_file.write(f"--- Segment {idx+1} ---\n")
            segmented_file.write(segment + "\n\n")

    print(f"Segmented text saved for {contract_name}")

    # Step 5: Create a standardized template from the segmented text
    standardized_template = create_standardized_template(segments)

    # Save the standardized template with the user-provided contract type
    template_file_path = os.path.join(contract_folder_path, f'{contract_type}_standardized_template.txt')
    with open(template_file_path, 'w') as template_file:
        template_file.write(standardized_template)

    print(f"Standardized template saved for {contract_name} as {contract_type}_standardized_template.txt")

    return template_file_path

# Function to segment the extracted text based on headings
def segment_text(text):
    # Define a regular expression pattern for detecting headings like Section, Article, Clause, etc.
    heading_pattern = r'\b(Section|Article|Clause|Chapter|Part)\b[\s:]+[\w\s]+'
    headings = re.findall(heading_pattern, text)

    # Split the text based on the headings
    segments = re.split(heading_pattern, text) if headings else [text.strip()]

    return [segment.strip() for segment in segments if segment.strip()]

# Function to create a standardized template from the segmented text
def create_standardized_template(segments):
    standardized_template = ""

    for idx, segment in enumerate(segments):
        standardized_template += f"--- Standardized Template for Section {idx+1} ---\n"
        standardized_template += f"Section/Article/Clause: Section {idx+1}\n"
        standardized_template += f"Content: {segment}\n\n"

    return standardized_template

# Example usage
pdf_path = "/content/drive/MyDrive/Business Contract Validation/input/distributor.pdf"  # Replace with actual path for input contract PDF
process_input_contract(pdf_path)
```
